```python
# These are all the modules we'll be using later. Make sure you can import them
# before proceeding further.
from __future__ import print_function
import os
import numpy as np
import random
import string
import tensorflow as tf
import zipfile
from six.moves import range
from six.moves.urllib.request import urlretrieve
```

```python
url = 'http://mattmahoney.net/dc/'

def maybe_download(filename, expected_bytes):
  """Download a file if not present, and make sure it's the right size."""
  if not os.path.exists(filename):
    filename, _ = urlretrieve(url + filename, filename)
  statinfo = os.stat(filename)
  if statinfo.st_size == expected_bytes:
    print('Found and verified %s' % filename)
  else:
    print(statinfo.st_size)
    raise Exception(
      'Failed to verify ' + filename + '. Can you get to it with a browser?')
  return filename

filename = maybe_download('text8.zip', 31344016)
```

```python
def read_data(filename):
  with zipfile.ZipFile(filename) as f:
    name = f.namelist()[0]
    data = tf.compat.as_str(f.read(name))
  return data
  
text = read_data(filename)
print('Data size %d' % len(text))
```

Utility functions to map characters to vocabulary IDs and back.

```python
vocabulary_size = len(string.ascii_lowercase) + 1 # [a-z] + ' '
first_letter = ord(string.ascii_lowercase[0])

def char2id(char):
  if char in string.ascii_lowercase:
    return ord(char) - first_letter + 1
  elif char == ' ':
    return 0
  else:
    print('Unexpected character: %s' % char)
    return 0
  
def id2char(dictid):
  if dictid > 0:
    return chr(dictid + first_letter - 1)
  else:
    return ' '

print(char2id('a'), char2id('z'), char2id(' '), char2id('ï'))
print(id2char(1), id2char(26), id2char(0))
```

Function to generate a training batch for the LSTM model.

```python
batch_size=64
num_unrollings=10

class BatchGenerator(object):
  def __init__(self, text, batch_size, num_unrollings):
    self._text = text
    self._text_size = len(text)
    self._batch_size = batch_size
    self._num_unrollings = num_unrollings
    segment = self._text_size // batch_size
    self._cursor = [ offset * segment for offset in range(batch_size)]
    self._last_batch = self._next_batch()
  
  def _next_batch(self):
    """Generate a single batch from the current cursor position in the data."""
    batch = np.zeros(shape=(self._batch_size, vocabulary_size), dtype=np.float)
    for b in range(self._batch_size):
      batch[b, char2id(self._text[self._cursor[b]])] = 1.0
      self._cursor[b] = (self._cursor[b] + 1) % self._text_size
    return batch
  
  def next(self):
    """Generate the next array of batches from the data. The array consists of
    the last batch of the previous array, followed by num_unrollings new ones.
    """
    batches = [self._last_batch]
    for step in range(self._num_unrollings):
      batches.append(self._next_batch())
    self._last_batch = batches[-1]
    return batches

def characters(probabilities):
  """Turn a 1-hot encoding or a probability distribution over the possible
  characters back into its (most likely) character representation."""
  return [id2char(c) for c in np.argmax(probabilities, 1)]

def batches2string(batches):
  """Convert a sequence of batches back into their (most likely) string
  representation."""
  s = [''] * batches[0].shape[0]
  for b in batches:
    s = [''.join(x) for x in zip(s, characters(b))]
  return s

train_batches = BatchGenerator(train_text, batch_size, num_unrollings)
valid_batches = BatchGenerator(valid_text, 1, 1)

print(batches2string(train_batches.next()))
print(batches2string(train_batches.next()))
print(batches2string(valid_batches.next()))
print(batches2string(valid_batches.next()))
```

```python
def logprob(predictions, labels):
  """Log-probability of the true labels in a predicted batch."""
  predictions[predictions < 1e-10] = 1e-10
  return np.sum(np.multiply(labels, -np.log(predictions))) / labels.shape[0]

def sample_distribution(distribution):
  """Sample one element from a distribution assumed to be an array of normalized
  probabilities.
  """
  r = random.uniform(0, 1)
  s = 0
  for i in range(len(distribution)):
    s += distribution[i]
    if s >= r:
      return i
  return len(distribution) - 1

def sample(prediction):
  """Turn a (column) prediction into 1-hot encoded samples."""
  p = np.zeros(shape=[1, vocabulary_size], dtype=np.float)
  p[0, sample_distribution(prediction[0])] = 1.0
  return p

def random_distribution():
  """Generate a random column of probabilities."""
  b = np.random.uniform(0.0, 1.0, size=[1, vocabulary_size])
  return b/np.sum(b, 1)[:,None]
```

Simple LSTM Model.

```python
num_nodes = 64

graph = tf.Graph()
with graph.as_default():
  
  # Parameters:
  # Input gate: input, previous output, and bias.
  ix = tf.Variable(tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
  im = tf.Variable(tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
  ib = tf.Variable(tf.zeros([1, num_nodes]))
  # Forget gate: input, previous output, and bias.
  fx = tf.Variable(tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
  fm = tf.Variable(tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
  fb = tf.Variable(tf.zeros([1, num_nodes]))
  # Memory cell: input, state and bias.                             
  cx = tf.Variable(tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
  cm = tf.Variable(tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
  cb = tf.Variable(tf.zeros([1, num_nodes]))
  # Output gate: input, previous output, and bias.
  ox = tf.Variable(tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
  om = tf.Variable(tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
  ob = tf.Variable(tf.zeros([1, num_nodes]))
  # Variables saving state across unrollings.
  saved_output = tf.Variable(tf.zeros([batch_size, num_nodes]), trainable=False)
  saved_state = tf.Variable(tf.zeros([batch_size, num_nodes]), trainable=False)
  # Classifier weights and biases.
  w = tf.Variable(tf.truncated_normal([num_nodes, vocabulary_size], -0.1, 0.1))
  b = tf.Variable(tf.zeros([vocabulary_size]))
  
  # Definition of the cell computation.
  def lstm_cell(i, o, state):
    """Create a LSTM cell. See e.g.: http://arxiv.org/pdf/1402.1128v1.pdf
    Note that in this formulation, we omit the various connections between the
    previous state and the gates."""
    print(i.shape)
    print(o.shape)
    print(state.shape)
    input_gate = tf.sigmoid(tf.matmul(i, ix) + tf.matmul(o, im) + ib)
    forget_gate = tf.sigmoid(tf.matmul(i, fx) + tf.matmul(o, fm) + fb)
    update = tf.matmul(i, cx) + tf.matmul(o, cm) + cb
    state = forget_gate * state + input_gate * tf.tanh(update)
    output_gate = tf.sigmoid(tf.matmul(i, ox) + tf.matmul(o, om) + ob)
    return output_gate * tf.tanh(state), state

  # Input data.
  train_data = list()
  for _ in range(num_unrollings + 1):
    train_data.append(
      tf.placeholder(tf.float32, shape=[batch_size,vocabulary_size]))
  train_inputs = train_data[:num_unrollings]
  train_labels = train_data[1:]  # labels are inputs shifted by one time step.

  # Unrolled LSTM loop.
  outputs = list()
  output = saved_output
  state = saved_state
  for i in train_inputs:
    output, state = lstm_cell(i, output, state)
    outputs.append(output)

  # State saving across unrollings.
  with tf.control_dependencies([saved_output.assign(output),
                                saved_state.assign(state)]):
    # Classifier.
    logits = tf.nn.xw_plus_b(tf.concat(outputs, 0), w, b)
    loss = tf.reduce_mean(
      tf.nn.softmax_cross_entropy_with_logits(
        labels=tf.concat(train_labels, 0), logits=logits))

  # Optimizer.
  global_step = tf.Variable(0)
  learning_rate = tf.train.exponential_decay(
    10.0, global_step, 5000, 0.1, staircase=True)
  optimizer = tf.train.GradientDescentOptimizer(learning_rate)
  gradients, v = zip(*optimizer.compute_gradients(loss))
  gradients, _ = tf.clip_by_global_norm(gradients, 1.25)
  optimizer = optimizer.apply_gradients(
    zip(gradients, v), global_step=global_step)

  # Predictions.
  train_prediction = tf.nn.softmax(logits)
  
  # Sampling and validation eval: batch 1, no unrolling.
  sample_input = tf.placeholder(tf.float32, shape=[1, vocabulary_size])
  saved_sample_output = tf.Variable(tf.zeros([1, num_nodes]))
  saved_sample_state = tf.Variable(tf.zeros([1, num_nodes]))
  reset_sample_state = tf.group(
    saved_sample_output.assign(tf.zeros([1, num_nodes])),
    saved_sample_state.assign(tf.zeros([1, num_nodes])))
  print('sample')
  sample_output, sample_state = lstm_cell(
    sample_input, saved_sample_output, saved_sample_state)
  with tf.control_dependencies([saved_sample_output.assign(sample_output),
                                saved_sample_state.assign(sample_state)]):
    sample_prediction = tf.nn.softmax(tf.nn.xw_plus_b(sample_output, w, b))
```

```python
num_steps = 7001
summary_frequency = 100

with tf.Session(graph=graph) as session:
  tf.global_variables_initializer().run()
  print('Initialized')
  mean_loss = 0
  for step in range(num_steps):
    batches = train_batches.next()
    feed_dict = dict()
    for i in range(num_unrollings + 1):
      feed_dict[train_data[i]] = batches[i]
    _, l, predictions, lr = session.run(
      [optimizer, loss, train_prediction, learning_rate], feed_dict=feed_dict)
    mean_loss += l
    if step % summary_frequency == 0:
      if step > 0:
        mean_loss = mean_loss / summary_frequency
      # The mean loss is an estimate of the loss over the last few batches.
      print(
        'Average loss at step %d: %f learning rate: %f' % (step, mean_loss, lr))
      mean_loss = 0
      labels = np.concatenate(list(batches)[1:])
      print('Minibatch perplexity: %.2f' % float(
        np.exp(logprob(predictions, labels))))
      if step % (summary_frequency * 10) == 0:
        # Generate some samples.
        print('=' * 80)
        for _ in range(5):
          feed = sample(random_distribution())
          sentence = characters(feed)[0]
          reset_sample_state.run()
          for _ in range(79):
            prediction = sample_prediction.eval({sample_input: feed})
            feed = sample(prediction)
            sentence += characters(feed)[0]
          print(sentence)
        print('=' * 80)
      # Measure validation set perplexity.
      reset_sample_state.run()
      valid_logprob = 0
      for _ in range(valid_size):
        b = valid_batches.next()
        predictions = sample_prediction.eval({sample_input: b[0]})
        valid_logprob = valid_logprob + logprob(predictions, b[1])
      print('Validation set perplexity: %.2f' % float(np.exp(
        valid_logprob / valid_size)))
```

```python
num_nodes = 64
num_steps = 7001
summary_frequency = 100


# Refactor rnn with LSTM cells
def prepare_params_for_rnn_with_lstm(
       vocabulary_size, 
       num_nodes,
    ):
    """
    Prepare params for rnn with lstm
    """
    # Parameters:
    # Input gate: input, previous output, and bias.
    input_gate_input = tf.Variable(
        tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
    input_gate_prev_output = tf.Variable(
        tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
    input_gate_bias = tf.Variable(tf.zeros([1, num_nodes]))
    
    # Forget gate: input, previous output, and bias.
    forget_gate_input = tf.Variable(
        tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
    forget_gate_prev_output = tf.Variable(
        tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
    forget_gate_bias = tf.Variable(tf.zeros([1, num_nodes]))
    
    # Memory cell: input, state and bias.                             
    memory_cell_input = tf.Variable(
        tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
    memory_cell_state = tf.Variable(
        tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
    memory_cell_bias = tf.Variable(tf.zeros([1, num_nodes]))
    
    # Output gate: input, previous output, and bias.
    output_gate_input = tf.Variable(
        tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
    output_gate_prev_output = tf.Variable(
        tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
    output_gate_bias = tf.Variable(tf.zeros([1, num_nodes]))
    
    return (
        input_gate_input,
        input_gate_prev_output,
        input_gate_bias,
        
        forget_gate_input,
        forget_gate_prev_output,
        forget_gate_bias,
        
        memory_cell_input,
        memory_cell_state,
        memory_cell_bias,
        
        output_gate_input,
        output_gate_prev_output,
        output_gate_bias)

def lstm_cell_with_diff_params_per_gate(
        i, o, state,
        input_gate_input,
        input_gate_prev_output,
        input_gate_bias,
        
        forget_gate_input,
        forget_gate_prev_output,
        forget_gate_bias,
        
        memory_cell_input,
        memory_cell_state,
        memory_cell_bias,
        
        output_gate_input,
        output_gate_prev_output,
        output_gate_bias):
    """Create a LSTM cell. See e.g.: http://arxiv.org/pdf/1402.1128v1.pdf
    Note that in this formulation, we omit the various connections between the
    previous state and the gates."""

    input_gate = tf.sigmoid(
        tf.matmul(i, input_gate_input) + \
        tf.matmul(o, input_gate_prev_output) + \
        input_gate_bias)
    forget_gate = tf.sigmoid(
        tf.matmul(i, forget_gate_input) + \
        tf.matmul(o, forget_gate_prev_output) + forget_gate_bias)
    update = \
        tf.matmul(i, memory_cell_input) + \
        tf.matmul(o, memory_cell_state) + \
        memory_cell_bias
    state = forget_gate * state + input_gate * tf.tanh(update)
    output_gate = tf.sigmoid(
        tf.matmul(i, output_gate_input) + \
        tf.matmul(o, output_gate_prev_output) + \
        output_gate_bias)
    return output_gate * tf.tanh(state), state


def get_output_for_rnn_with_lstm(
        train_inputs,
    
        cell_function,

        saved_output,
        saved_state,

        *cell_params):
    """
    Get output for rnn with lstm cells
    """
    
    # Unrolled LSTM loop.
    outputs = list()
    output = saved_output
    state = saved_state
    for i in train_inputs:
        output, state = cell_function(
            i, output, state, 
            *cell_params)
        outputs.append(output)   
    return tf.concat(outputs, 0), output, state

def generate_rnn_graph_with_lstm(
       vocabulary_size, 
       num_nodes,
       batch_size,
       num_unrollings,
    
       prepare_params_for_rnn_with_lstm,
       cell_function,
       get_output_for_rnn_with_lstm
    ):
    """
    Generate TF graph for rnn with LSTM cells
    """
    gate_params = prepare_params_for_rnn_with_lstm(
        vocabulary_size, 
        num_nodes,)
    
    # Variables saving state across unrollings.
    saved_output = tf.Variable(tf.zeros([batch_size, num_nodes]), trainable=False)
    saved_state = tf.Variable(tf.zeros([batch_size, num_nodes]), trainable=False)

        # Classifier weights and biases.
    w = tf.Variable(tf.truncated_normal([num_nodes, vocabulary_size], -0.1, 0.1))
    b = tf.Variable(tf.zeros([vocabulary_size]))

    # Input data.
    train_data = list()
    for _ in range(num_unrollings + 1):
        train_data.append(
            tf.placeholder(tf.float32, shape=[batch_size, vocabulary_size]))
        train_inputs = train_data[:num_unrollings]
        train_labels = train_data[1:]  # labels are inputs shifted by one time step.

    print(
        'Train inputs shape', len(train_inputs), train_inputs[0].shape)
    print(
        'Train labels shape', len(train_labels), train_labels[0].shape)

    outputs, last_output, state = get_output_for_rnn_with_lstm(
        train_inputs,
        
        cell_function,

        saved_output,
        saved_state,

        *gate_params)
    
    print('Outputs shape', outputs.shape)
    print('Last output shape', last_output.shape)
    print('State shape', state.shape)

    # State saving across unrollings.
    with tf.control_dependencies([
            saved_output.assign(last_output),
            saved_state.assign(state)]):
        # Classifier.
        logits = tf.nn.xw_plus_b(outputs, w, b)
        loss = tf.reduce_mean(
        tf.nn.softmax_cross_entropy_with_logits(
            labels=tf.concat(train_labels, 0), logits=logits))

    # Optimizer.
    global_step = tf.Variable(0)
    learning_rate = tf.train.exponential_decay(
       10.0, global_step, 5000, 0.1, staircase=True)
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    gradients, v = zip(*optimizer.compute_gradients(loss))
    gradients, _ = tf.clip_by_global_norm(gradients, 1.25)
    optimizer = optimizer.apply_gradients(
       zip(gradients, v), global_step=global_step)

    # Predictions.
    train_prediction = tf.nn.softmax(logits)
    print('Train prediction shape', train_prediction.shape)
  
    # Sampling and validation eval: batch 1, no unrolling.
    sample_input = tf.placeholder(tf.float32, shape=[1, vocabulary_size])
    saved_sample_output = tf.Variable(tf.zeros([1, num_nodes]))
    saved_sample_state = tf.Variable(tf.zeros([1, num_nodes]))
    reset_sample_state = tf.group(
        saved_sample_output.assign(tf.zeros([1, num_nodes])),
        saved_sample_state.assign(tf.zeros([1, num_nodes])))
    
    sample_outputs, sample_last_output, sample_state = get_output_for_rnn_with_lstm(
        [sample_input, ],
        
        cell_function,

        saved_sample_output, 
        saved_sample_state,
    
        *gate_params)
    with tf.control_dependencies(
        [
            saved_sample_output.assign(sample_last_output),
            saved_sample_state.assign(sample_state)
        ]):
        sample_prediction = tf.nn.softmax(tf.nn.xw_plus_b(sample_last_output, w, b))
        
    session_args = [
        optimizer, 
        loss, 
        train_prediction, 
        learning_rate]
    return (
        session_args,
        
        sample_prediction,
        sample_input,
        reset_sample_state,
        
        train_data,
    )


def calculate_minibatch_perplexity(
        predictions, batches
    ):
    labels = np.concatenate(list(batches)[1:])
    print('Minibatch perplexity: %.2f' % float(
        np.exp(logprob(predictions, labels))))


def run_session_for_rnn_with_lstm_cells(
        session,
        train_batches,
        valid_batches,
    
        session_args,
    
        sample_prediction,
        sample_input,
        reset_sample_state,

        summary_frequency,
        valid_size,
        num_unrollings,
    
        train_data,
     
        generate_and_print_samples,
        measure_validation_set_perplexity,
        calculate_minibatch_perplexity,
    
        vocabulary_size
    ):
    """
    Execute tf graph in session to train and evaluate rnn with lstm cells
    """
    def _run_step(step):
        batches = train_batches.next()
        feed_dict = {
            train_data[i]: batches[i]
            for i in range(num_unrollings + 1)
        }
        
        if step == 0:
            print('Train data shape', len(train_data), train_data[0].shape)
            print('Batches example', batches)

        session_res = session.run(
           session_args, feed_dict=feed_dict)
        _, l, predictions, lr  = session_res
       
        if step % summary_frequency == 0 and step > 0:
            calculate_minibatch_perplexity(
                predictions, batches)
        return l, lr
    
    
    tf.global_variables_initializer().run()
    print('Initialized')
    mean_loss = 0
    for step in range(num_steps):
        l, lr = _run_step(step)
        mean_loss += l
    
        if step % summary_frequency == 0 and step > 0:
            mean_loss = mean_loss / summary_frequency
    
            print(
                'Average loss at step %d: %.2f Learning rate: %.2f' % (step, mean_loss, lr))
            mean_loss = 0

        if step % (summary_frequency * 10) == 0:
            # Generate some samples.
            print('=' * 80)
            generate_and_print_samples(
                reset_sample_state,
                sample_prediction,
                sample_input)
            print('=' * 80)

            # Measure validation set perplexity.
            reset_sample_state.run()
            measure_validation_set_perplexity(
                valid_size,
                valid_batches,
                sample_prediction,
                sample_input,
                vocabulary_size)

            
def measure_validation_set_perplexity(
        valid_size,
        valid_batches,
        sample_prediction,
        sample_input,
        *args
    ):
    valid_logprob = 0
    for _ in range(valid_size):
        b = valid_batches.next()
        predictions = sample_prediction.eval({sample_input: b[0]})
        valid_logprob += logprob(predictions, b[1])
    print(
        'Validation set perplexity: %.2f' % float(
         np.exp(valid_logprob / valid_size)))


def generate_and_print_samples(
        reset_sample_state,
        sample_prediction,
        sample_input):
    for _ in range(5):
        feed = sample(random_distribution())
        sentence = characters(feed)[0]
        reset_sample_state.run()
        for _ in range(79):
            prediction = sample_prediction.eval({sample_input: feed})
            feed = sample(prediction)
            sentence += characters(feed)[0]
        print(sentence)

    
def run_rnn_with_lstm_model(
        vocabulary_size, 
        num_nodes,
        batch_size,
        num_unrollings,
        valid_size,
        summary_frequency,
    
        train_batches,
        valid_batches,
    
        prepare_params_for_rnn_with_lstm,
        cell_function,
        get_output_for_rnn_with_lstm,
    
        generate_and_print_samples,
        measure_validation_set_perplexity,
        calculate_minibatch_perplexity
    ):
    """
    Train and evaluate rnn model with lstm cells
    """
    tf.reset_default_graph()
    graph = tf.Graph()
    with graph.as_default(), tf.device('/cpu:0'):
        (
            session_args,
        
            sample_prediction,
            sample_input,
            reset_sample_state,

            train_data,           
        ) = generate_rnn_graph_with_lstm(
            vocabulary_size, 
            num_nodes,
            batch_size,
            num_unrollings,

            prepare_params_for_rnn_with_lstm,
            cell_function,
            get_output_for_rnn_with_lstm
        )
        
    with tf.Session(graph=graph) as session:
        run_session_for_rnn_with_lstm_cells(
            session,
            train_batches,
            valid_batches,
    
            session_args,
    
            sample_prediction,
            sample_input,
            reset_sample_state,

            summary_frequency,
            valid_size,
            num_unrollings,
            train_data,

            generate_and_print_samples,
            measure_validation_set_perplexity,
            calculate_minibatch_perplexity,
        
            vocabulary_size)

        
run_rnn_with_lstm_model(
        vocabulary_size, 
        num_nodes,
        batch_size,
        num_unrollings,
        valid_size,
        summary_frequency,
    
        BatchGenerator(train_text, batch_size, num_unrollings),
        BatchGenerator(valid_text, 1, 1),

        prepare_params_for_rnn_with_lstm,
        lstm_cell_with_diff_params_per_gate,
        get_output_for_rnn_with_lstm,

        generate_and_print_samples,
        measure_validation_set_perplexity,
        calculate_minibatch_perplexity)
```

Problem 1
---------

You might have noticed that the definition of the LSTM cell involves 4 matrix multiplications with the input, and 4 matrix multiplications with the output. Simplify the expression by using a single matrix multiply for each, and variables that are 4 times larger.

---

```python
# Calculate LSTM value with 4 times less matmuls

GATES_AND_CELLS_NUMBER = 4 # 3 gates and 1 memeory cell

def prepare_params_for_rnn_with_fast_lstm(
       vocabulary_size, 
       num_nodes,
    ):
    """
    Prepare params for rnn with lstm
    """
    # Parameters:
    gates_input = tf.Variable(
        tf.truncated_normal(
            [
                vocabulary_size, 
                num_nodes * GATES_AND_CELLS_NUMBER # input gate + forget gate + memory cell + output gate
            ], -0.1, 0.1))
    gates_prev_output = tf.Variable(
        tf.truncated_normal(
            [
                num_nodes, 
                num_nodes * GATES_AND_CELLS_NUMBER # input gate + forget gate + memory cell + output gate
            ], -0.1, 0.1))

    input_gate_bias = tf.Variable(tf.zeros([1, num_nodes]))
    forget_gate_bias = tf.Variable(tf.zeros([1, num_nodes]))
    memory_cell_bias = tf.Variable(tf.zeros([1, num_nodes]))
    output_gate_bias = tf.Variable(tf.zeros([1, num_nodes]))
    
    return (
        gates_input,
        gates_prev_output,
        
        input_gate_bias,
        forget_gate_bias,
        memory_cell_bias,
        output_gate_bias)


def fast_lstm_cell(
        i, o, state,

        gates_input,
        gates_prev_output,
        
        input_gate_bias,
        forget_gate_bias,
        memory_cell_bias,
        output_gate_bias):
    """Create a LSTM cell using only 2 matmuls: for input and output"""

    i = tf.matmul(i, gates_input)
    o = tf.matmul(o, gates_prev_output)
    (
        input_gate_i,
        forget_gate_i,
        memory_cell_i,
        output_gate_i
    ) = tf.split(i, GATES_AND_CELLS_NUMBER, axis=1)
    (
        input_gate_o,
        forget_gate_o,
        memory_cell_o,
        output_gate_o
    ) = tf.split(o, GATES_AND_CELLS_NUMBER, axis=1)

    input_gate = tf.sigmoid(
        input_gate_i + \
        input_gate_o + \
        input_gate_bias)
    forget_gate = tf.sigmoid(
        forget_gate_i + \
        forget_gate_o + forget_gate_bias)
    update = \
        memory_cell_i + \
        memory_cell_o + \
        memory_cell_bias
    state = forget_gate * state + input_gate * tf.tanh(update)
    output_gate = tf.sigmoid(
        output_gate_i + \
        output_gate_o + \
        output_gate_bias)
    return output_gate * tf.tanh(state), state

run_rnn_with_lstm_model(
        vocabulary_size, 
        num_nodes,
        batch_size,
        num_unrollings,
        valid_size,
        summary_frequency,
    
        BatchGenerator(train_text, batch_size, num_unrollings),
        BatchGenerator(valid_text, 1, 1),

        prepare_params_for_rnn_with_fast_lstm,
        fast_lstm_cell,
        get_output_for_rnn_with_lstm,

        generate_and_print_samples,
        measure_validation_set_perplexity,
        calculate_minibatch_perplexity)
```

We want to train a LSTM over bigrams, that is pairs of consecutive characters like 'ab' instead of single characters like 'a'. Since the number of possible bigrams is large, feeding them directly to the LSTM using 1-hot encodings will lead to a very sparse representation that is very wasteful computationally.

a- Introduce an embedding lookup on the inputs, and feed the embeddings to the LSTM cell instead of the inputs themselves.

b- Write a bigram-based LSTM, modeled on the character LSTM above.

c- Introduce Dropout. For best practices on how to use Dropout in LSTMs, refer to this [article](http://arxiv.org/abs/1409.2329).

```python
import math

# Feed embeddings instead of the inputs
class BatchGeneratorWithoutOneHotEncoding(
      BatchGenerator):
    """
    Generates batches from text without using one hot encoding
    """
    def __init__(self, *args, **kwargs):
        self._cursor_no_unrolling = 0
        super(
            BatchGeneratorWithoutOneHotEncoding, self).__init__(*args, **kwargs)
    
    def _next_batch(self):
        """Generate a single batch from the current cursor position in the data."""
        batch = np.zeros(shape=(self._batch_size,), dtype=np.int32)
        for b in range(self._batch_size):
            batch[b] = char2id(self._text[self._cursor[b]])
            self._cursor[b] = (self._cursor[b] + 1) % self._text_size            
        return batch

EMBEDDINGS_SIZE = 27
batch_size = 64
num_unrollings = 10
num_nodes = 64

def generate_rnn_graph_with_lstm_and_embeddings(
       embeddings_size, 
       vocabulary_size,
       num_nodes,
       batch_size,
       num_unrollings,
    
       prepare_params_for_rnn_with_lstm,
       cell_function,
       get_output_for_rnn_with_lstm
    ):
    """
    Generate TF graph for rnn with LSTM cells using embeddings
    """
    gate_params = prepare_params_for_rnn_with_lstm(
        embeddings_size, 
        num_nodes,)
    
    # Variables saving state across unrollings.
    saved_output = tf.Variable(tf.zeros([batch_size, num_nodes]), trainable=False)
    saved_state = tf.Variable(tf.zeros([batch_size, num_nodes]), trainable=False)

    # Classifier weights and biases.
    w = tf.Variable(tf.truncated_normal([num_nodes, vocabulary_size]))
                                        #stddev=1.0 / math.sqrt(embeddings_size)))
    sampled_w = tf.transpose(w)
    b = tf.Variable(tf.zeros([vocabulary_size]))

    # Embeddings
    embeddings = tf.Variable(
        tf.random_uniform([vocabulary_size, embeddings_size], -1.0, 1.0))
        
    # Input data.
    train_data = list()
    train_labels_input = list()
    for _ in range(num_unrollings + 1):
        train_dataset = tf.placeholder(
            tf.int32, shape=[batch_size, ])
        train_data.append(train_dataset)
        train_label = tf.placeholder(
            tf.int32, shape=[batch_size, vocabulary_size])
        train_labels_input.append(train_label)
        #embed_dataset = tf.nn.embedding_lookup(embeddings, train_dataset)
        #embed_data.append(embed_dataset)

    train_inputs = tf.concat(train_data, 0)#embed_data[:num_unrollings]
    embed_inputs = tf.nn.embedding_lookup(embeddings, train_inputs)
    train_inputs = tf.split(
        embed_inputs, num_or_size_splits=embed_inputs.shape[0] // batch_size, axis=0)
    train_inputs_embed = train_inputs[:num_unrollings]

    train_labels = train_labels_input[1:]  # labels are inputs shifted by one time step.
    train_labels = tf.concat(train_labels, 0)
    
    train_labels_embed = tf.concat(train_inputs[1:], 0)
    

    print(
        'Train inputs shape', len(train_inputs), train_inputs[0].shape)
    print(
        'Train labels shape', train_labels.shape)

    outputs, last_output, state = get_output_for_rnn_with_lstm(
        train_inputs_embed,
        
        cell_function,

        saved_output,
        saved_state,

        *gate_params)
    
    print('Outputs shape', outputs.shape)
    print('Last output shape', last_output.shape)
    print('State shape', state.shape)

    # State saving across unrollings.
    with tf.control_dependencies([
            saved_output.assign(last_output),
            saved_state.assign(state)]):
        # Classifier.
        logits = tf.nn.xw_plus_b(outputs, w, b)
        loss = tf.reduce_mean(
            tf.nn.softmax_cross_entropy_with_logits(
                labels=train_labels, logits=logits))

    # Optimizer.
    global_step = tf.Variable(0)
    learning_rate = tf.train.exponential_decay(
       10.0, global_step, 5000, 0.1, staircase=True)
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    gradients, v = zip(*optimizer.compute_gradients(loss))
    gradients, _ = tf.clip_by_global_norm(gradients, 1.10)
    optimizer = optimizer.apply_gradients(
       zip(gradients, v), global_step=global_step)

    #optimizer = tf.train.GradientDescentOptimizer(1).minimize(loss)

    # Predictions.
    train_prediction = tf.nn.softmax(logits)
    print('Train prediction shape', train_prediction.shape)
  
    # Sampling and validation eval: batch 1, no unrolling.
    sample_input = tf.placeholder(tf.int32, shape=[1, ])
    sample_embed = tf.nn.embedding_lookup(embeddings, sample_input)
    saved_sample_output = tf.Variable(tf.zeros([1, num_nodes]))
    saved_sample_state = tf.Variable(tf.zeros([1, num_nodes]))
    reset_sample_state = tf.group(
        saved_sample_output.assign(tf.zeros([1, num_nodes])),
        saved_sample_state.assign(tf.zeros([1, num_nodes])))
    
    sample_outputs, sample_last_output, sample_state = get_output_for_rnn_with_lstm(
        [sample_embed, ],
        
        cell_function,

        saved_sample_output, 
        saved_sample_state,
    
        *gate_params)
    with tf.control_dependencies(
        [
            saved_sample_output.assign(sample_last_output),
            saved_sample_state.assign(sample_state)
        ]):
        sample_prediction = tf.nn.softmax(tf.nn.xw_plus_b(sample_last_output, w, b))
        print('Sample prediction shape', sample_prediction.shape)

    session_args = [
        optimizer, 
        loss, 
        train_prediction,
        learning_rate] 
    
    norm = tf.sqrt(tf.reduce_sum(tf.square(embeddings), 1, keepdims=True))
    normalized_embeddings = embeddings / norm
    return (
        session_args,
        
        sample_prediction,
        sample_input,
        reset_sample_state,
        
        train_data,
        train_labels_input,
        embeddings,
        normalized_embeddings,

        train_prediction
    )


def run_session_for_rnn_with_lstm_cells_with_embeddings(
    session,
    train_batches,
    train_label_batches,
    valid_batches,
    valid_labels_batches,
    
    session_args,
    
    sample_prediction,
    sample_input,
    reset_sample_state,

    summary_frequency,
    valid_size,
    num_unrollings,
    train_data,
    train_labels,
        
    generate_and_print_samples,
    measure_validation_set_perplexity,
    calculate_minibatch_perplexity,
        
    vocabulary_size,

    embeddings,
    normalized_embeddings,

    train_prediction):
    """
    Run session for word2vec model
    """
    def _run_step(step):
        batches = train_batches.next()
        labels = train_label_batches.next()
        feed_dict = {}
        for i in range(num_unrollings + 1):
            feed_dict[train_data[i]] = batches[i]
            feed_dict[train_labels[i]] = labels[i]
        
        if step == 0:
            print('Train data shape', len(train_data), train_data[0].shape)
            #print('Batches example', batches)
            #print('Labels example', labels)
        #if step % (summary_frequency * 10) == 0:
            #print('Train predictions for step %d' % step, train_prediction.eval(feed_dict))

        session_res = session.run(
           session_args, feed_dict=feed_dict)
        _, l, predictions, lr = session_res
       
        if (step % summary_frequency == 0) and step > 0:
            calculate_minibatch_perplexity(
                predictions, labels)
        return l, lr
    
    
    tf.global_variables_initializer().run()
    print('Initialized')
    mean_loss = 0
    for step in range(num_steps):
        l, lr = _run_step(step)
        mean_loss += l
    
        if step % summary_frequency == 0 and step > 0:
            mean_loss = mean_loss / summary_frequency
    
            print(
                'Average loss at step %d: %.2f. Learning rate: %.2f' % (step, mean_loss, lr))
            mean_loss = 0

        if step % (summary_frequency * 10) == 0 and step > 0:
            # Generate some samples.
            print('=' * 80)
            generate_and_print_samples(
                reset_sample_state,
                sample_prediction,
                sample_input)
            print('=' * 80)

            # Measure validation set perplexity.
            reset_sample_state.run()
            measure_validation_set_perplexity(
                valid_size,
                valid_batches,
                valid_labels_batches,
                sample_prediction,
                sample_input,
                vocabulary_size)


def run_rnn_with_lstm_model_and_embeddings(
        vocabulary_size, 
        embeddings_size,
        num_nodes,
        batch_size,
        num_unrollings,
        valid_size,
        summary_frequency,
    
        train_batches,
        train_label_batches,
        valid_batches,
        valid_labels_batches,
    
        prepare_params_for_rnn_with_lstm,
        cell_function,
        get_output_for_rnn_with_lstm,
    
        generate_and_print_samples,
        measure_validation_set_perplexity,
        calculate_minibatch_perplexity
    ):
    """
    Train and evaluate rnn model with lstm cells
    """
    tf.reset_default_graph()
    graph = tf.Graph()
    with graph.as_default(), tf.device('/cpu:0'):
        (
            session_args,
            sample_prediction,
            sample_input,
            reset_sample_state,

            train_data, 
            train_labels,
            embeddings,
            normalized_embeddings,
            train_prediction
        ) = generate_rnn_graph_with_lstm_and_embeddings(
            embeddings_size,
            vocabulary_size, 
            num_nodes,
            batch_size,
            num_unrollings,

            prepare_params_for_rnn_with_lstm,
            cell_function,
            get_output_for_rnn_with_lstm
        )
        
    with tf.Session(graph=graph) as session:
        run_session_for_rnn_with_lstm_cells_with_embeddings(
            session,
            train_batches,
            train_label_batches,
            valid_batches,
            valid_labels_batches,
    
            session_args,
    
            sample_prediction,
            sample_input,
            reset_sample_state,

            summary_frequency,
            valid_size,
            num_unrollings,
            train_data,
            train_labels,
        
            generate_and_print_samples,
            measure_validation_set_perplexity,
            calculate_minibatch_perplexity,
        
            vocabulary_size,
            embeddings,
            normalized_embeddings,
            train_prediction)


def generate_and_print_samples_for_embeddings(
        reset_sample_state,
        sample_prediction,
        sample_input):
    for _ in range(5):
        feed = random.randint(0, vocabulary_size - 1)
        sentence = id2char(feed)
        reset_sample_state.run()
        for _ in range(79):
            prediction = sample_prediction.eval({sample_input: [feed, ]})
            feed = prediction.argsort(axis=1)[0][-1]
            sentence += id2char(feed)
        print(sentence)

 
def measure_validation_set_perplexity_with_embeddings(
        valid_size,
        valid_batches,
        valid_labels_batches,
        sample_prediction,
        sample_input,
        vocabulary_size,
    ):
    valid_logprob = 0
    for _ in range(valid_size):
        b = valid_batches.next()
        predictions = sample_prediction.eval({sample_input: b[0]})
        true_value = np.concatenate(valid_labels_batches.next()[1:], 0)
        valid_logprob += logprob(predictions, true_value)
    print(
        'Validation set perplexity: %.2f' % float(
         np.exp(valid_logprob / valid_size)))

    
def calculate_minibatch_perplexity_with_embeddings(
        predictions, batches
    ):
    labels = np.concatenate(list(batches)[1:])
    print('Minibatch perplexity: %.2f' % float(
        np.exp(logprob(predictions, labels))))
    

run_rnn_with_lstm_model_and_embeddings(
        vocabulary_size, 
        EMBEDDINGS_SIZE,
        num_nodes,
        batch_size,
        num_unrollings,
        valid_size,
        summary_frequency,
    
        BatchGeneratorWithoutOneHotEncoding(
            train_text, batch_size, num_unrollings),
        BatchGenerator(
            train_text, batch_size, num_unrollings),
        BatchGeneratorWithoutOneHotEncoding(
            valid_text, 1, 1),
        BatchGenerator(
            valid_text, 1, 1),

        prepare_params_for_rnn_with_fast_lstm,
        fast_lstm_cell,
        get_output_for_rnn_with_lstm,

        generate_and_print_samples_for_embeddings,
        measure_validation_set_perplexity_with_embeddings,
        calculate_minibatch_perplexity_with_embeddings)
```

```python
# LSTM with bigrams
BIGRAMS_VOCABULARY_SIZE = vocabulary_size * vocabulary_size
BIGRAMS_EMBEDDINGS_SIZE = 128
BIGRAMS_NUM_NODES = 64
BIGRAMS_BATCH_SIZE = 10
BIGRAMS_NUM_UNROLLINGS = 61


def bigram2id(bigram):
    """
    Genetrate id from bigram
    """
    return char2id(bigram[0]) * vocabulary_size + char2id(bigram[1])


def id2bigram(id_):
    """
    Generate bigram from id
    """
    return ''.join([
            id2char(id_ // vocabulary_size),
            id2char(id_ % vocabulary_size)
        ])


class BigramsBatchGeneratorWithoutOneHotEncoding(
        BatchGeneratorWithoutOneHotEncoding):
    """
    Generates bigrams batches without one hot encoding
    """    
    def _next_batch(self):
        """Generate a single batch from the current cursor position in the data."""
        batch = np.zeros(shape=(self._batch_size,), dtype=np.int32)
        for b in range(self._batch_size):
            bigram = self._text[self._cursor[b]: self._cursor[b] + 2]
            if len(bigram) < 2:
                bigram += ' '
            batch[b] = bigram2id(bigram)
            self._cursor[b] = (self._cursor[b] + 2) % self._text_size            
        return batch


class BigramsBatchGenerator(BatchGenerator):
    """
    Generates bigrams batches with one hot encoding
    """
  
    def _next_batch(self):
        """Generate a single batch from the current cursor position in the data."""
        batch = np.zeros(shape=(self._batch_size, BIGRAMS_VOCABULARY_SIZE), dtype=np.float)
        for b in range(self._batch_size):
            bigram = self._text[self._cursor[b]: self._cursor[b] + 2]
            if len(bigram) < 2:
                bigram += ' '
            batch[b, bigram2id(bigram)] = 1.0
            self._cursor[b] = (self._cursor[b] + 2) % self._text_size
        return batch


def generate_and_print_samples_for_bigramms(
        reset_sample_state,
        sample_prediction,
        sample_input):
    for _ in range(5):
        feed = random.randint(0, BIGRAMS_VOCABULARY_SIZE - 1)
        sentence = id2bigram(feed)
        reset_sample_state.run()
        for _ in range(79):
            prediction = sample_prediction.eval({sample_input: [feed, ]})
            feed = prediction.argsort(axis=1)[0][-1]
            sentence += id2bigram(feed)
        print(sentence)


run_rnn_with_lstm_model_and_embeddings(
        BIGRAMS_VOCABULARY_SIZE, 
        BIGRAMS_EMBEDDINGS_SIZE,
        BIGRAMS_NUM_NODES,
        BIGRAMS_BATCH_SIZE,
        BIGRAMS_NUM_UNROLLINGS,
        valid_size,
        summary_frequency,
    
        BigramsBatchGeneratorWithoutOneHotEncoding(
            train_text, 
            BIGRAMS_BATCH_SIZE, BIGRAMS_NUM_UNROLLINGS),
        BigramsBatchGenerator(
            train_text, 
            BIGRAMS_BATCH_SIZE, BIGRAMS_NUM_UNROLLINGS),
        BigramsBatchGeneratorWithoutOneHotEncoding(
            valid_text, 1, 1),
        BigramsBatchGenerator(
            valid_text, 1, 1),

        prepare_params_for_rnn_with_fast_lstm,
        fast_lstm_cell,
        get_output_for_rnn_with_lstm,

        generate_and_print_samples_for_bigramms,
        measure_validation_set_perplexity_with_embeddings,
        calculate_minibatch_perplexity_with_embeddings)
```

```python
# Introduce Dropouts

DROPOUT_KEEP_PROBABILITY = 0.95

def get_output_for_rnn_with_lstm_and_dropouts(
        train_inputs,
    
        cell_function,

        saved_output,
        saved_state,

        *cell_params):
    """
    Get output for rnn with lstm cells.
    Apply dropout to the input
    """
    train_inputs_with_dropouts = []
    for train_input in train_inputs:
        train_inputs_with_dropouts.append(
            tf.nn.dropout(
                train_input,
                DROPOUT_KEEP_PROBABILITY))
        
    return get_output_for_rnn_with_lstm(
        train_inputs_with_dropouts,
    
        cell_function,

        saved_output,
        saved_state,

        *cell_params)


run_rnn_with_lstm_model_and_embeddings(
        BIGRAMS_VOCABULARY_SIZE, 
        BIGRAMS_EMBEDDINGS_SIZE,
        BIGRAMS_NUM_NODES,
        BIGRAMS_BATCH_SIZE,
        BIGRAMS_NUM_UNROLLINGS,
        valid_size,
        summary_frequency,
    
        BigramsBatchGeneratorWithoutOneHotEncoding(
            train_text, 
            BIGRAMS_BATCH_SIZE, BIGRAMS_NUM_UNROLLINGS),
        BigramsBatchGenerator(
            train_text, 
            BIGRAMS_BATCH_SIZE, BIGRAMS_NUM_UNROLLINGS),
        BigramsBatchGeneratorWithoutOneHotEncoding(
            valid_text, 1, 1),
        BigramsBatchGenerator(
            valid_text, 1, 1),

        prepare_params_for_rnn_with_fast_lstm,
        fast_lstm_cell,
        get_output_for_rnn_with_lstm_and_dropouts,

        generate_and_print_samples_for_bigramms,
        measure_validation_set_perplexity_with_embeddings,
        calculate_minibatch_perplexity_with_embeddings)
```
